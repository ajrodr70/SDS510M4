```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
```

```python
##Examine news.asu.edu website
thispage = requests.get('https://news.asu.edu/')
bs = BeautifulSoup(thispage.text,'html.parser')
paragraphs = bs.find_all('a', href=True)
news_links = []
for paragraph in paragraphs:
  if 'news.asu.edu' in paragraph['href']:
    news_links.append(paragraph['href'])
    print(paragraph['href'], "\n")
print(len(news_links))
```

```python
##Use list(set) to get unique URLs
news_links = list(set(news_links))
#print(len(news_links))

##Create DataFrame
df = pd.DataFrame(columns=['url', 'keyword'])

##Look for keyword within paragraphs of websites from new_links variable
for article in news_links:
 article_html = requests.get(article)
 soup = BeautifulSoup(article_html.text, 'html.parser')
 sleep(1)
 for p_element in soup.find_all('p'):
  if 'medicine' in p_element.get_text():
    medicine_url = {'url': article, 'keyword': 'medicine'}
    df.loc[len(df)] = medicine_url
  if 'breakthrough' in p_element.get_text():
    breakthrough_url = {'url': article, 'keyword': 'breakthrough'}
    df.loc[len(df)] = breakthrough_url
  if 'safe' in p_element.get_text():
    safe_url = {'url': article, 'keyword': 'safe'}
    df.loc[len(df)] = safe_url
  if 'teacher' in p_element.get_text():
    teacher_url = {'url': article, 'keyword': 'teacher'}
    df.loc[len(df)] = teacher_url

print(df)

```

```python
#Create column within DataFrame to remove duplicate rows
concat_id = []

for index in range(0, len(df)):
  concat_url = df['url'][index] + df['keyword'][index]
  concat_id.append(concat_url)

df['concat_id'] = concat_id

##Drop duplicate rows found with column concat_id, then remove column concat_id, leaving url and keyword column
df = df.drop_duplicates(subset=['concat_id'], ignore_index=True)
df = df.drop(['concat_id'], axis=1)
print(df)
```

```python
df.to_csv('news_links.csv')
```
